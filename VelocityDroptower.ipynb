```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.integrate import cumtrapz

vdata=np.loadtxt('droptower_vdata.txt',usecols=(0,1))
time = np.array(vdata)[:,0]
vel = np.array(vdata)[:,1]

position = cumtrapz(vel,time,initial=0) + 30
acceleration = np.diff(vel) #time interval is one sec
plt.title('Drop Tower')
plt.plot(time,position,label = 'position')
plt.plot(time,vel,label='velocity')
plt.plot(time[0:10],acceleration,label='acceleration')
plt.legend()
```

```python
pos_accel = [x for x in acceleration if x>0]
mean_accel = np.mean(pos_accel)
print(mean_accel)
```
